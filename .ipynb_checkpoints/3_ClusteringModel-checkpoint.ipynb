# Executive Summary
In this following section of the project, our aim was to make use of clustering techniques to segment the customers into groups so that the telco company may identify customer segments and come up with targeted strategies to target and attract more of those customers. 

Since our dataset consist of both numerical and categorical data, we have decided to use the KPrototype clustering technique. 

Before clustering, we need to identify the optimal number of clusters for clustering. We used the Elbow method and Average Silhouette Method. We will elaborate on both methods below.

After getting the optimal number of clusters, we are ready to use the Kprototype to cluster our data.

After clustering, we needed to do cluster interpretation. We used a summary data table that shows a summary of the (mean and mode of the) features of each cluster in order to get an idea of what each cluster is about.

We realised that it is very difficult to identify and profile our clusters when there are so many features to it. Hence, we need a way to pick out just a few important variables. 

In order to identify important variables for cluster profiling, we have used the Shapley plot. The Shapley plot ranks the variables in order of importance in forming the various clusters. From there, we identified and chose the top 3 most important variables to help us with cluster profiling. 

Lastly, we did an analysis on the clusters.

### Below, we map the categorical data to numbers

```python
##https://pbpython.com/categorical-encoding.html
cleanup_nums = {"gender":     {"Female": 1, "Male": 0},
                "partner": {"Yes": 1, "No": 0},
                "dependents": {"Yes": 1, "No": 0},
                "phoneservice": {"Yes": 1, "No": 0},
                "multiplelines": {"Yes": 1, "No phone service": 0,"No": 2},
                "internetservice": {"DSL": 1, "No": 0,"Fiber optic": 2},
                "onlinesecurity": {"Yes": 1, "No internet service": 0,"No": 2},
                "onlinebackup": {"Yes": 1, "No internet service": 0,"No": 2},
                "deviceprotection": {"Yes": 1, "No internet service": 0,"No": 2},
                "techsupport": {"Yes": 1, "No internet service": 0,"No": 2},
                "streamingtv": {"Yes": 1, "No internet service": 0,"No": 2},
                "streamingmovies": {"Yes": 1, "No internet service": 0,"No": 2},
                "contract": {"Month-to-month": 1, "One year": 0,"Two year": 2},
                "paperlessbilling": {"Yes": 1, "No": 0},
                "paymentmethod": {"Electronic check": 1, "Mailed check": 0,
                                  "Bank transfer (automatic)": 2, "Credit card (automatic)": 3}
                
               }


df_scale = df_scale.replace(cleanup_nums)
df_scale.head()

```

### Before clustering, we need to identify the optimal number of clusters for clustering.

### To do so, we first used the Elbow method as it is the most common method for finding optimal clusters.


```python
#https://python.plainenglish.io/data-science-project-clustering-mixed-data-7d5fd6e7f047
# Get the position of categorical columns
categorical_columns = ['gender', 'seniorcitizen','partner', 'dependents','phoneservice', 'multiplelines','internetservice', 'onlinesecurity','onlinebackup', 'deviceprotection','techsupport', 'streamingtv', 'streamingmovies','contract', 'paperlessbilling','paymentmethod']
catColumnsPos = []
for c in categorical_columns:
    catColumnsPos.append(df_scale.columns.get_loc(c))
catColumnsPos
```

```python
#https://towardsdatascience.com/the-k-prototype-as-clustering-algorithm-for-mixed-data-type-categorical-and-numerical-fe7c50538ebb
# Convert dataframe to matrix. It helps the kmodes module running the K-Prototype clustering algorithm. 
dfMatrix = df_scale.to_numpy()
dfMatrix
```

### From the plot below, we want to find the elbow point which is the point/cluster number where the value of the cost is not rapidly changing anymore. This elbow point will tell us the optimal number of clusters.

### We see that the elbow point lies at around 5 clusters.

```python
##https://python.plainenglish.io/data-science-project-clustering-mixed-data-7d5fd6e7f047

##Note: Will take an extremely long time to load and run 

#Choose optimal number of clusters using Elbow method
import seaborn as sns
num_clusters = list(range(2, 10))
cost_values = []
# calculate cost values for each number of clusters (2 to 10)
for k in num_clusters:
    kproto = KPrototypes(n_clusters=k, init='Huang', random_state=42)
    kproto.fit_predict(dfMatrix, categorical= catColumnsPos)
    cost_values.append(kproto.cost_)
# plot cost against number of clusters
ax = sns.lineplot(x=num_clusters, y=cost_values, marker="o")
ax.set_title('Elbow curve', fontsize=14)
ax.set_xlabel('No of clusters', fontsize=11)
ax.set_ylabel('Cost', fontsize=11)
```

### Next, we tried a different method for determining the optimal number of cluster, such as using the Average Silhouette Method. A silhouette value of 1 indicates the data point is well clustered. (i.e. far away from other clustered points). We want to look for the number of clusters that result in the maximum average silhouette value.

### From the plot below, we see that 3 clusters result in the largest average silhouette score within the range of 2 to 10. 

### Now from both cluster evaluation methods, we will go with 4 clusters, the middle of 3 and 5.

```python
##https://python.plainenglish.io/data-science-project-clustering-mixed-data-7d5fd6e7f047
##Note: Will take an extremely long time to load and run 

from sklearn.metrics import silhouette_score
num_clusters = list(range(2, 10))
silhouette_avg = []
# calculate average silhouette score for each number of cluster (2 to 10)
for k in num_clusters:
    kproto = KPrototypes(n_clusters=k, init='Huang', random_state=42)
    kproto.fit_predict(dfMatrix, categorical= catColumnsPos)
    cluster_labels = kproto.labels_
    silhouette_avg.append(silhouette_score(dfMatrix, cluster_labels))
# plot average silhouette score against number of clusters
ax = sns.lineplot(x=num_clusters, y=silhouette_avg, marker="o")
ax.set_title('Average Silhouette', fontsize=14)
ax.set_xlabel('No of clusters', fontsize=11)
ax.set_ylabel('score', fontsize=11)
```

### First Model used:
### Applying the K-prototype clustering method to cluster the data below.

```python
##https://python.plainenglish.io/data-science-project-clustering-mixed-data-7d5fd6e7f047
# we set the number of clusters to 4
kproto = KPrototypes(n_clusters= 4, init='Huang', n_init = 25, random_state=42)
kproto.fit_predict(dfMatrix, categorical= catColumnsPos)
```

```python
##https://python.plainenglish.io/data-science-project-clustering-mixed-data-7d5fd6e7f047
##https://towardsdatascience.com/the-k-prototype-as-clustering-algorithm-for-mixed-data-type-categorical-and-numerical-fe7c50538ebb

# store cluster labels
cluster_labels = kproto.labels_
# add clusters to dataframe
Totaldata["ClusterLabel"] = cluster_labels

Totaldata["ClusterSegment"] = Totaldata["ClusterLabel"].map({0:'One', 1:'Two', 2:'Three',3:"Four"})

cleanup_nums = {"ClusterLabel":{0: 1, 1: 2,2: 3,3:4} }
Totaldata = Totaldata.replace(cleanup_nums)

Totaldata["ClusterLabel"]= Totaldata["ClusterLabel"].astype('category')
Totaldata
```

### After we have the clusters, we want to look at the statistics for each cluster so that we can interpret the clusters.
### Below is the summary table for each clusters.

#### Second model used: 
#### Before we have the Shapley plot, we used the LGBMClassifier classfier model based on the clusters and evaluated the classification model using the f1 score. 

```python
proto_labs = kproto.labels_
```

```python
clf_kp = LGBMClassifier(colsample_by_tree=0.8)
cv_scores_kp = cross_val_score(clf_kp, cat_data, proto_labs, scoring='f1_weighted') #score help to train and test together
print(f'CV F1 score for K-Prototypes clusters is {np.mean(cv_scores_kp)}') #accuracy score
```

```python
clf_kp.fit(cat_data, proto_labs) #fit model
```

### Third model used: Shapley.

### The Shapley plot ranks the variables in order of importance and it helps us figure out the top features in forming the various clusters. The top 5 variables are TotalCharges, MonthlyCharges, tenure, InternetService, and StreamingTV.

```python
explainer_kp = shap.TreeExplainer(clf_kp)
shap_values_kp = explainer_kp.shap_values(cat_data) 
```

```python
shap.summary_plot(shap_values_kp, cat_data, plot_type="bar", plot_size=(15, 10))
```

## Cluster Interpretation and Data Driven Insights and Recommendations

```python
summary = Totaldata.groupby('ClusterSegment').agg(
    {
        'Count':'count',
        'totalcharges': 'mean',
        'monthlycharges': 'mean',
        'tenure': 'mean',
        'internetservice': lambda x: x.value_counts().index[0],
        'streamingtv': lambda x: x.value_counts().index[0],
        
    }
).reset_index()

summary

```

We can profile the clusters using these 5 variables: TotalCharges, MonthlyCharges, Tenure, InternetService and StreamingTV

We know that customers can either subscribe to Internet Service only, or Phone Service only or both Internet Service and Phone Service.

For cluster 1, it has the lowest total charges and monthly charges. As these group of customers generally did not subscribe to the company's internet services, it means that they are only with the company due to their Phone services contract. This might explain why their total spendings is the lowest among the 4 clusters. Thus, We can say this cluster consists of low spender customers. For cluster 1, lowest spenders who only subscribe to phone services, telco company can promote their internet services to them or offer bundles consisting of both internet and phone services to these customers. This will help attract this group to subscribe to the internet too.

For cluster 2, it has the highest total charges and monthly charges. These group of customers generally subsribe only to the internet and Streaming TV service. With the longest tenure and highest spendings, we can say that they are the loyal and high spenders of the company. For cluster 2, the high spenders who subscribes to most services, we feel that it would not be so much of increasing their spendings, but more of deepening the relationship with these customers. Hence, the company can offer special services and discounts to these customers to show appreciation for their loyalty over the years.

As for cluster 3, its the largest cluster. Its impt to note that it has the lowest tenure. Upon a closer look at the main summary table, we see that this cluster subscribes to both internet and phone services. However, they generally do not subscribe to any of the add on services. Thus, we think that this cluster consist of the average spenders customers. 
For cluster 3, the average spenders who only subscribes to main services only, Telco can offer more promotions on their add on services to these customers. 

For cluster 4, the spendings of this cluster is similar to cluster 2, but the difference between cluster 2 and 3 is that this group generally has the lower tenure even though their total charges are higher. We believe that this might be attributed to their subscription of Internet Services and streaming TV. Thus, We have named this cluster as high but volatile customers. 
For Cluster 4, the high spenders who generally has the lower tenure even though their spendings are higher, Telco company can come up with customer loyalty programs to ensure these group of customers will come back and renew their contract with the company in the future.
